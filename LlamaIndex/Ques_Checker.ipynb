```python
from llama_index.llms.groq import Groq
```

```python
from dotenv import load_dotenv
import os 
load_dotenv()

## load the Groq API key
groq_api_key = os.environ['GROQ_API_KEY']
```

```python
llm = Groq(model="mixtral-8x7b-32768", api_key=groq_api_key)
```

```python
from llama_index.core.llms import ChatMessage
answer="tuples are mutable data struc in python"
messages = [
    ChatMessage(
        role="system", content="this is a question which user is going to answer - explain tuples in python"
    ),
    ChatMessage(role="user", content=f"give marks to {answer}/{"5"} out of 5, dont write anything else only one number should exist in output absolutely no explaination ."),
]
resp = llm.chat(messages)
print(resp)
```
